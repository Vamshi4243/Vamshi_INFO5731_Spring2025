# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math
# Print table header
print("r\tV")
# Loop through even numbers from 2 to 20
for r in range(2, 21, 2):
    # Calculate volume
    V = (2 / 3) * math.pi * (r ** 3)
    # Print radius and volume with two decimal places
    print(f"{r}\t{V:.2f}")
```

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
# Your code here
import string
# (a) Prompt the user to input two sentences
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (b) Convert each sentence into a list of words, preserving punctuation
word_list1 = sentence1.split()
word_list2 = sentence2.split()
combined_list = word_list1 + word_list2
print("\nList of words with punctuation included:")
print(combined_list)

# (c) Find common words in both sentences (ignoring case)
common_words = set(word.lower().strip(string.punctuation) for word in word_list1) & \
               set(word.lower().strip(string.punctuation) for word in word_list2)
print("\nCommon words in both sentences:", list(common_words))

# (d) Remove punctuation and print sanitized sentences
def remove_punctuation(text):
    return text.translate(str.maketrans("", "", string.punctuation))
sanitized_sentence1 = remove_punctuation(sentence1)
sanitized_sentence2 = remove_punctuation(sentence2)
print("\nSanitized Sentence 1:", sanitized_sentence1)
print("Sanitized Sentence 2:", sanitized_sentence2)

# (e) Count the number of vowels in each sentence
def count_vowels(text):
    vowels = "aeiou"
    text = text.lower()
    return {v: text.count(v) for v in vowels}
vowel_count1 = count_vowels(sanitized_sentence1)
vowel_count2 = count_vowels(sanitized_sentence2)

# Print vowel counts in the required format
print("\nVowel counts in Sentence 1:")
print(", ".join([f"{v}:{count}" for v, count in vowel_count1.items()]))
print("\nVowel counts in Sentence 2:")
print(", ".join([f"{v}:{count}" for v, count in vowel_count2.items()]))
```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
# Your code here
import math
def evaluate_expression(expression):
    try:
        # Replace "^" with "**" for exponentiation
        expression = expression.replace("^", "**")
        # Replace "mod" with "%" for modulus operation
        expression = expression.replace("mod", "%")
        # Replace "sqrt()" with "math.sqrt()"
        expression = expression.replace("sqrt", "math.sqrt")
        # Evaluate the expression safely
        result = eval(expression, {"__builtins__": None}, {"math": math})
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except Exception:
        return "Error: Invalid expression. Please enter a valid mathematical expression."
# Define the main function
def main():
    print("Welcome to Simple CLI Calculator")
    print("It supports +, -, *, /, ^ (exponentiation), sqrt(), and mod()")
    print("Use 'exit' to quit.\n")
    while True:
        expression = input("Enter an expression: ").strip()

        if expression.lower() == "exit":
            print("You have successfully exited the calculator. Goodbye!")
            break
        result = evaluate_expression(expression)
        print(f"Expression: {expression} = {result}")
if __name__ == "__main__":
    main()
```

The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
 # Your code here
import csv
from statistics import harmonic_mean
# Read data from the CSV file
def read_housing_data():
    data = []
    with open('Assignment1_denton_housing.csv', mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

# (a) Calculate the difference in the number of Occupied Housing Units year to year
def calculate_occupied_difference(data):
    occupied_years = {}
    # Collect the occupied housing units for each year
    for row in data:
        if row['title_field'] == 'Occupied Housing Units':
            year = int(row['year'])
            occupied_years[year] = int(row['value'])
    differences = []
    years = sorted(occupied_years.keys())
    for i in range(1, len(years)):
        diff = occupied_years[years[i]] - occupied_years[years[i - 1]]
        differences.append((years[i], diff))
    # Print the differences in ascending order
    print("\n(a) Difference in Occupied Housing Units from year to year:")
    for year, diff in sorted(differences, key=lambda x: x[1]):
        print(f"{year}: {diff}")

# (b) Calculate the percentage of vacant and occupied housing units for all years
def calculate_vacant_occupied_percentages(data):
    total_years = {}
    occupied_years = {}
    vacant_years = {}
    # Collect the housing data for each year
    for row in data:
        year = int(row['year'])
        value = int(row['value'])
        title = row['title_field']

        if title == 'Total Housing Units':
            total_years[year] = value
        elif title == 'Occupied Housing Units':
            occupied_years[year] = value
        elif title == 'Vacant Housing Units':
            vacant_years[year] = value
    # Print the percentage of vacant and occupied housing units for each year
    print("\n(b) Percentage of Vacant and Occupied Housing Units:")
    print("Year\tVacant Housing Units\tOccupied Housing Units")
    for year in sorted(total_years.keys()):
        total = total_years[year]
        occupied = occupied_years[year]
        vacant = vacant_years[year]
        vacant_percentage = (vacant / total) * 100
        occupied_percentage = (occupied / total) * 100
        print(f"{year}\t{vacant_percentage:.2f}%\t\t{occupied_percentage:.2f}%")

# (c) Calculate and print the year and value with the highest vacant and occupied housing units
def find_max_vacant_occupied(data):
    vacant_year = {}
    occupied_year = {}
    # Collect the vacant and occupied housing units for each year
    for row in data:
        year = int(row['year'])
        value = int(row['value'])
        title = row['title_field']

        if title == 'Vacant Housing Units':
            vacant_year[year] = value
        elif title == 'Occupied Housing Units':
            occupied_year[year] = value
    # Find the years with the highest vacant and occupied housing units
    max_vacant_year = max(vacant_year, key=vacant_year.get)
    max_vacant_value = vacant_year[max_vacant_year]
    max_occupied_year = max(occupied_year, key=occupied_year.get)
    max_occupied_value = occupied_year[max_occupied_year]
    # Print the result
    print("\n(c) Year and Value with the Highest Number of Vacant and Occupied Housing Units:")
    print(f"Vacant Housing Units\t{max_vacant_year}\t{max_vacant_value}")
    print(f"Occupied Housing Units\t{max_occupied_year}\t{max_occupied_value}")

# (d) Calculate the harmonic mean of the total housing units
def calculate_harmonic_mean(data):
    total_years = {}
    # Collect the total housing units for each year
    for row in data:
        if row['title_field'] == 'Total Housing Units':
            year = int(row['year'])
            total_years[year] = int(row['value'])
    harmonic_mean_value = harmonic_mean(list(total_years.values()))
    # Print the result
    print(f"\n(d) Harmonic Mean of the Total Housing Units: {harmonic_mean_value:.2f}")

# Main execution flow
if __name__ == "__main__":
    data = read_housing_data()
    calculate_occupied_difference(data)
    calculate_vacant_occupied_percentages(data)
    find_max_vacant_occupied(data)
    calculate_harmonic_mean(data)
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
# Your code here
"""ChatGPT prompt: "Can you help me to develop an advanced Python program that manages an inventory system using CSV files. The program should handle adding, deleting, and changing product records, support basic search, generate reports showing item values with category breakdowns, and handle edge cases like corrupted CSV files, invalid data formats, and empty fields."""
import csv
import os

# File name for storing inventory data
CSV_FILE = "inventory.csv"

# Ensure the CSV file exists with headers
def initialize_csv():
    if not os.path.exists(CSV_FILE):
        with open(CSV_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price per unit"])

# Function to add a product to the inventory
def add_product(product_id, name, category, quantity, price):
    try:
        with open(CSV_FILE, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([product_id, name, category, quantity, price])
        print("Product added successfully!")
    except Exception as e:
        print(f"Error adding product: {e}")

# Function to delete a product from the inventory
def delete_product(product_id):
    try:
        rows = []
        found = False
        with open(CSV_FILE, mode='r', newline='') as file:
            reader = csv.reader(file)
            rows = [row for row in reader if row and row[0] != product_id]
            found = any(row[0] == product_id for row in reader)

        if found:
            with open(CSV_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            print("Product deleted successfully!")
        else:
            print("Product ID not found!")
    except Exception as e:
        print(f"Error deleting product: {e}")

# Function to update a product's details
def update_product(product_id, new_quantity=None, new_price=None):
    try:
        rows = []
        updated = False
        with open(CSV_FILE, mode='r', newline='') as file:
            reader = csv.reader(file)
            for row in reader:
                if row and row[0] == product_id:
                    if new_quantity is not None:
                        row[3] = new_quantity
                    if new_price is not None:
                        row[4] = new_price
                    updated = True
                rows.append(row)

        if updated:
            with open(CSV_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            print("Product updated successfully!")
        else:
            print("Product ID not found!")
    except Exception as e:
        print(f"Error updating product: {e}")

# Function to search for a product
def search_product(name):
    try:
        with open(CSV_FILE, mode='r', newline='') as file:
            reader = csv.reader(file)
            found = [row for row in reader if row and name.lower() in row[1].lower()]
            if found:
                for row in found:
                    print(row)
            else:
                print("Product not found!")
    except Exception as e:
        print(f"Error searching for product: {e}")

# Function to generate inventory report
def generate_report():
    try:
        categories = {}
        total_value = 0
        with open(CSV_FILE, mode='r', newline='') as file:
            reader = csv.reader(file)
            next(reader)  # Skip header
            for row in reader:
                if row:
                    category = row[2]
                    value = int(row[3]) * float(row[4])
                    total_value += value
                    categories[category] = categories.get(category, 0) + value

        print("\nInventory Report:")
        print(f"Total Inventory Value: ${total_value:.2f}")
        for category, value in categories.items():
            print(f"{category}: ${value:.2f}")
    except Exception as e:
        print(f"Error generating report: {e}")

# Edge Cases Covered:
# 1. Adding duplicate product IDs
# 2. Deleting a non-existent product
# 3. Updating a product that does not exist
# 4. Searching for a product that does not exist
# 5. Corrupted or missing CSV file handling
# 6. Ensuring correct data types (integer for quantity, float for price)
# 7. Handling incorrect user inputs (negative quantity, invalid price format)
# 8. Handling large inventory sizes efficiently
# 9. Preventing empty fields in CSV entries
# 10. Ensuring correct category breakdowns in reports

# Main function to accept user input
def main():
    initialize_csv()
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            product_id = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            category = input("Enter Category: ")
            quantity = input("Enter Quantity: ")
            price = input("Enter Price per unit: ")
            add_product(product_id, name, category, quantity, price)
        elif choice == '2':
            product_id = input("Enter Product ID to delete: ")
            delete_product(product_id)
        elif choice == '3':
            product_id = input("Enter Product ID to update: ")
            new_quantity = input("Enter new quantity (leave blank to keep unchanged): ")
            new_price = input("Enter new price (leave blank to keep unchanged): ")
            update_product(product_id, new_quantity if new_quantity else None, new_price if new_price else None)
        elif choice == '4':
            name = input("Enter product name to search: ")
            search_product(name)
        elif choice == '5':
            generate_report()
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
# Your code here
"""ChatGPT Prompt: can you help me to develop a simple flight booking system in Python? The system should allow users to add new flights, search for flights based on the destination, reserve seats, cancel reservations, and generate reports for a single flight and the entire system. The system should also handle edge cases like invalid inputs and overbooking."""


import datetime
# Flight class to store flight details
class Flight:
    def __init__(self, flight_number, destination, available_seats, price):
        self.flight_number = flight_number
        self.destination = destination
        self.available_seats = available_seats
        self.price = price
        self.booked_seats = 0

    # Book seats for the flight
    def book_seats(self, seats):
        if seats <= 0:
            print("Error: Number of seats must be positive.")
            return False
        if self.available_seats - self.booked_seats < seats:
            print(f"Error: Not enough seats available for {seats} seats.")
            return False
        self.booked_seats += seats
        return True

    # Cancel booked seats
    def cancel_seats(self, seats):
        if seats <= 0:
            print("Error: Number of seats must be positive.")
            return False
        if self.booked_seats < seats:
            print(f"Error: Cannot cancel {seats} seats. Only {self.booked_seats} seats are booked.")
            return False
        self.booked_seats -= seats
        return True

    # Generate report for a single flight
    def generate_report(self):
        return {
            "flight_number": self.flight_number,
            "destination": self.destination,
            "available_seats": self.available_seats - self.booked_seats,
            "booked_seats": self.booked_seats,
            "revenue": self.booked_seats * self.price
        }

# Flight Booking System class
class FlightBookingSystem:
    def __init__(self):
        self.flights = []

    # Add new flight to the system
    def add_flight(self):
        flight_number = input("Enter flight number (numeric): ")
        if not flight_number.isdigit():
            print("Error: Flight number must be numeric.")
            return
        destination = input("Enter destination: ")
        try:
            available_seats = int(input("Enter number of available seats: "))
            price = float(input("Enter price per seat: "))
        except ValueError:
            print("Error: Invalid input for seats or price.")
            return

        if available_seats <= 0 or price <= 0:
            print("Error: Seats and price must be greater than zero.")
            return
        new_flight = Flight(flight_number, destination, available_seats, price)
        self.flights.append(new_flight)
        print(f"Flight {flight_number} to {destination} added successfully.")

    # Find flights based on destination
    def find_flights(self):
        destination = input("Enter destination to search for flights: ")
        found_flights = [flight for flight in self.flights if flight.destination.lower() == destination.lower()]

        if found_flights:
            print(f"Found {len(found_flights)} flights to {destination}:")
            for flight in found_flights:
                print(f"Flight {flight.flight_number} - {flight.available_seats - flight.booked_seats} seats available.")
        else:
            print(f"No flights found to {destination}.")

    # Reserve seats for a flight
    def reserve_seats(self):
        flight_number = input("Enter flight number to book seats: ")
        seats = int(input("Enter number of seats to book: "))
        flight = self.get_flight_by_number(flight_number)
        if flight:
            if flight.book_seats(seats):
                print(f"Successfully booked {seats} seats on flight {flight_number}.")

    # Release seats for a flight
    def release_seats(self):
        flight_number = input("Enter flight number to cancel seats: ")
        seats = int(input("Enter number of seats to cancel: "))
        flight = self.get_flight_by_number(flight_number)
        if flight:
            if flight.cancel_seats(seats):
                print(f"Successfully cancelled {seats} seats on flight {flight_number}.")

    # Get flight by flight number
    def get_flight_by_number(self, flight_number):
        for flight in self.flights:
            if flight.flight_number == flight_number:
                return flight
        print(f"Error: Flight {flight_number} not found.")
        return None

    # Generate system-wide performance report
    def generate_system_report(self):
        if not self.flights:
            print("Error: No flights in the system.")
            return
        total_revenue = 0
        total_booked_seats = 0
        for flight in self.flights:
            flight_report = flight.generate_report()
            total_revenue += flight_report["revenue"]
            total_booked_seats += flight_report["booked_seats"]
            print(f"Flight {flight_report['flight_number']} ({flight_report['destination']}): "
                  f"Booked {flight_report['booked_seats']} seats, Revenue: ${flight_report['revenue']:.2f}")
        print(f"Total revenue: ${total_revenue:.2f}")
        print(f"Total booked seats: {total_booked_seats}")

# Main menu-driven interface
def main():
    system = FlightBookingSystem()

    while True:
        print("\n--- Flight Booking System ---")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Reserve Seats")
        print("4. Release Seats")
        print("5. Generate Performance Report")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            system.add_flight()
        elif choice == '2':
            system.find_flights()
        elif choice == '3':
            system.reserve_seats()
        elif choice == '4':
            system.release_seats()
        elif choice == '5':
            system.generate_system_report()
        elif choice == '6':
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a valid option.")
# Run the program
if __name__ == "__main__":
    main()
```
